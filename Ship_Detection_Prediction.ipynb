```python
import requests
import cv2
import numpy as np
from PIL import Image
from io import BytesIO
from ultralytics import YOLO

# Function to load an image from a URL
def load_image_from_url(url):
    response = requests.get(url)
    if response.status_code == 200:
        image = Image.open(BytesIO(response.content)).convert("RGB")
        return image
    else:
        raise Exception(f"Failed to load image from URL: {url}")

# Load the YOLO model
model = YOLO('best.onnx', task='detect')
```

```python
# Define the URL of the image
image_url = 'https://www.preligens.com/sites/default/files/2021-07/Ships.jpeg'

try:
    # Load and resize the image
    image = load_image_from_url(image_url)
    image = image.resize((640, 640))

    # Perform prediction using YOLO
    results = model.predict(image, conf=0.25)

    # Display the results
    results[0].show()  # Show the image with bounding boxes
except Exception as e:
    print(f"Error: {e}")
```
